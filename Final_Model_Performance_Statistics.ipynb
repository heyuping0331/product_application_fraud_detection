```python
import os
import pandas as pd
import numpy as np
```

# Load Train/Test/OOT Predictions One at a Time
## Columns: Record #, True label, Prediction (probability)

```python
#newDF = pd.read_csv("testprediction.csv")
#newDF = pd.read_csv("trainprediction.csv")
newDF = pd.read_csv("ootprediction.csv")
```

```python
newDF.sort_values(by=['prediction'], ascending = False, inplace = True)
```

# Calculate Model Performance Statistics

```python
# List initiation
fdr = []
records_cum = []
goods_cum = []
bads_cum = []
perc_goods = []
perc_bads = []

total_bads = sum(newDF['fraud_label']==1)
total_goods = sum(newDF['fraud_label']==0)
total_records = len(newDF)
fraud_rate = total_bads / total_records
```

```python
# Calculate stats from each cut-off

for i in np.arange(0.01, 0.21, 0.01):
    topRows = int(round(len(newDF)*i))
    records_cum.append(topRows)
    bads_cum.append(newDF.iloc[0:topRows, 2].sum())
    goods_cum.append(topRows - newDF.iloc[0:topRows, 2].sum())
    perc_goods.append((topRows - newDF.iloc[0:topRows, 2].sum())/ total_goods)
    fdr.append(newDF.iloc[0:topRows, 2].sum()/sum(newDF['fraud_label']==1))
    

print(total_bads, total_goods, total_records, fraud_rate)
```

```python
table_stats = pd.DataFrame({'fdr':fdr,
                          'records_cum':records_cum,
                          'goods_cum':goods_cum,
                          'bads_cum':bads_cum,
                          'perc_goods': perc_goods})
table_stats
```

```python
table_stats.to_excel('OOT stats.xlsx')
```
